```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://lhr.nu.edu.pk/fsc/faculty/'
r = requests.get(URL)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
#print(soup.prettify())
```

```python
id=[]
name=[]
desig=[]
bool_hec=[]
education=[]
email=[]
department=[]
extension=[]
link_img=[]
link_profile=[]
```

```python
def lhr_scrapper(all_cs,passed):
    for i in range(len(all_cs)):
        faculty = all_cs[i]
        try:
            id.append(int(faculty.find('a', class_ = 'faculty-link')['href'].split('/')[-1]))
        except:
            id.append(-1)
        name.append(faculty.find('h5', class_ = 'text-center').text)
        desig.append(faculty.find('p', class_ = 'small').text.strip().split('\n')[0])
        bool_hec.append('HEC' in (faculty.find('p', class_ = 'small').text.strip()))
        email.append(faculty.find('p', class_ = 'mb-0').text)
        link_img.append('https://lhr.nu.edu.pk' + faculty.find('img', class_ = 'card-img-top')['src'])
        link_profile.append('https://lhr.nu.edu.pk' + faculty.find('a', class_ = 'faculty-link')['href'])
        URL_temp =link_profile[-1]
        r_temp= requests.get(URL_temp)
        soup_temp= BeautifulSoup(r_temp.content, 'html.parser')
        soup2=soup_temp.find('div',class_='card-body')
        extension.append(soup2.find('p',class_='mt-0').text.split(':')[-1])
        try:
            ed=soup_temp.find('div',class_='col-lg-8')
            ed=ed.find('ul')
            ed=ed.findAll('li')[0].text.split('(')[0].replace(',','').replace(' ','')
        except:
            ed='None'
        education.append(ed)
        department.append(passed)
```

```python
all_comp=soup.findAll('div',{"id":'faculty-tab'})
all_cs=all_comp[0]
all_cs=all_cs.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_cs,'Computer Science')

all_ds=all_comp[1]
all_ds=all_ds.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_ds,'Data Science')

all_se=all_comp[2]
all_se=all_se.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_se,'Software Engineering')


URL_other= 'https://lhr.nu.edu.pk/faculty/'
r_other= requests.get(URL_other)
soup_other= BeautifulSoup(r_other.content, 'html.parser')
ee=soup_other.find('div',{"id":'ee'})
all_ee=ee.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_ee,'Electrical Engineering')

cv=soup_other.find('div',{"id":'cv'})
all_cv=cv.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_cv,'Civil Engineering')

fsm=soup_other.find('div',{"id":'fsm'})
all_fsm=fsm.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_fsm,'Management')

ss=soup_other.find('div',{"id":'ss'})
all_ss=ss.findAll('div', class_ ='facultyCard')
lhr_scrapper(all_ss,'Science and Humanities')
```

```python
df = pd.DataFrame({'ID':id,
                  'Name':name,
                  'Designation':desig,
                   'HEC Approved':bool_hec,
                   'Highest Ed.':education,
                  'Email':email,
                   'Department':department,
                   'Extension':extension,
                  'Profile':link_profile,
                  'Image':link_img})

df.to_csv('NUCES_FACULTY.csv')
```
